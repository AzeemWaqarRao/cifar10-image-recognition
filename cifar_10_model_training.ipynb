```python
from keras.layers import Dense,MaxPooling2D,Dropout,Flatten
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from keras.models import Sequential, Model, load_model
from keras.applications import InceptionV3
```

```python
inception = InceptionV3(include_top=False, input_shape=[128,128,3], weights='imagenet')
```

```python
for layer in inception.layers:
  layer.trainable = False
```

```python
x = Flatten()(inception.output)
prediction = Dense(64, activation='relu')(x)
prediction = Dense(32, activation='relu')(prediction)
prediction = Dropout(0.3)(prediction)
prediction = Dense(10,activation = "softmax")(prediction)
```

```python
model = Model(inputs=inception.input, outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', metrics=['accuracy'],loss='categorical_crossentropy')
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255, shear_range=0.2,zoom_range=0.2)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_data = train_datagen.flow_from_directory('/content/cifar10-128/train',batch_size=16,class_mode='categorical',target_size=(128,128))
```

```python
test_data = test_datagen.flow_from_directory('/content/cifar10-128/test',batch_size=16,class_mode='categorical',target_size=(128,128))
```

```python
model.fit_generator(train_data, epochs = 10, validation_data = test_data, shuffle = True)
```

```python
model.save('inception.h5')
```
